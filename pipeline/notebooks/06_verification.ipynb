This notebook verifies active spatial mining classifications using confusion matrix equations (Precision, Recall, F1-Score) and tests spatial dependency using Moran's I formulas.

```python
# 1. Confusion Matrix Calculations
tp = 88
fp = 6
fn = 12
tn = 94

precision = tp / (tp + fp)
recall = tp / (tp + fn)
f1 = 2 * (precision * recall) / (precision + recall)

print("Spatial Classification Metrics:")
print(f"- Precision: {precision:.4f}")
print(f"- Recall:    {recall:.4f}")
print(f"- F1-Score:  {f1:.4f}")
```

## Spatial Autocorrelation (Moran's I)
Moran's $I$ evaluates whether spatial anomalies are clustered, dispersed, or random. A value close to $+1.0$ indicates high spatial clustering (matching physical mining excavations).

```python
import numpy as np

# Create mock spatial coordinate anomalies to simulate Moran's I
np.random.seed(42)
size = 10
spatial_grid = np.random.normal(0, 1, (size, size))
# Create a clustered anomaly block in the center (representing mining cuts)
spatial_grid[3:7, 3:7] += 3.0

def calculate_morans_i(grid):
    N = grid.size
    mean = np.mean(grid)
    denominator = np.sum((grid - mean) ** 2)
    
    num_sum = 0.0
    weight_sum = 0.0
    
    # Calculate spatial weight matrix (rook contiguity)
    rows, cols = grid.shape
    for r in range(rows):
        for c in range(cols):
            val_i = grid[r, c]
            # Check 4 neighbors
            neighbors = []
            if r > 0: neighbors.append((r-1, c))
            if r < rows-1: neighbors.append((r+1, c))
            if c > 0: neighbors.append((r, c-1))
            if c < cols-1: neighbors.append((r, c+1))
            
            for nr, nc in neighbors:
                val_j = grid[nr, nc]
                num_sum += (val_i - mean) * (val_j - mean)
                weight_sum += 1.0
                
    moran_i = (N / weight_sum) * (num_sum / denominator)
    return moran_i

moran_val = calculate_morans_i(spatial_grid)
print(f"Calculated Moran's I spatial correlation: {moran_val:.4f}")
print(f"Spatial Clustering Confirmed (Moran's I > 0): {moran_val > 0}")
```
