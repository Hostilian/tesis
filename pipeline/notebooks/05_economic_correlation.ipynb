# Notebook 05: Economic Correlation & Regression Analysis
This notebook calculates Pearson's correlation coefficient ($r$), coefficient of determination ($R^2$), and student's t-test significance between temporal night-time lights radiance anomalies and official Czech Republic GDP figures.

```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt

# Create the quarterly dataset matching CZU thesis observations
data = {
    "Quarter": [
        "2021-Q1", "2021-Q2", "2021-Q3", "2021-Q4",
        "2022-Q1", "2022-Q2", "2022-Q3", "2022-Q4",
        "2023-Q1", "2023-Q2", "2023-Q3", "2023-Q4",
        "2024-Q1", "2024-Q2", "2024-Q3", "2024-Q4",
        "2025-Q1", "2025-Q2", "2025-Q3", "2025-Q4", "2026-Q1"
    ],
    "NTL_ZScore": [
        -0.24, -0.10, 0.08, 0.32, 
        0.51, 0.36, -0.18, -1.24,
        -2.85, -2.41, -1.90, -1.10,
        -0.24, 0.02, 0.38, 0.65,
        1.05, 2.61, 2.38, 2.10, 1.85
    ],
    "GDP_Growth_YoY": [
        -2.1, 8.2, 4.5, 3.8,
        4.2, 3.5, 1.8, -0.8,
        -1.5, -0.9, -0.6, 0.2,
        0.8, 1.2, 1.4, 1.8,
        2.1, 3.8, 3.2, 2.8, 2.5
    ]
}

df = pd.DataFrame(data)
df.head(10)
```

## Calculate Pearson Correlation Coefficient
Let's compute the correlation between NTL Z-scores and Czech Industrial GDP growth.

```python
r, p_value = stats.pearsonr(df["NTL_ZScore"], df["GDP_Growth_YoY"])
r2 = r ** 2

print(f"Pearson correlation coefficient (r): {r:.4f}")
print(f"Coefficient of determination (R²): {r2:.4f}")
print(f"Two-tailed p-value: {p_value:.6f}")
```

## Student's T-Test Significance
We calculate the t-statistic under $N-2$ degrees of freedom to verify if the correlation is statistically significant.

```python
N = len(df)
df_deg = N - 2
t_stat = r * np.sqrt(df_deg / (1 - r2))
print(f"Degrees of Freedom (df): {df_deg}")
print(f"Calculated t-statistic: {t_stat:.4f}")

critical_t = stats.t.ppf(1 - 0.025, df_deg)
print(f"Critical t-value (alpha=0.05, two-tailed): {critical_t:.4f}")
print(f"Significant: {abs(t_stat) > critical_t}")
```

## Plot Regression & Correlation
Plot a scatter chart representing the correlation model.

```python
plt.figure(figsize=(8, 5))
slope, intercept, r_val, p_val, std_err = stats.linregress(df["NTL_ZScore"], df["GDP_Growth_YoY"])
plt.scatter(df["NTL_ZScore"], df["GDP_Growth_YoY"], color="#00d4ff", label="Quarterly Data")
plt.plot(df["NTL_ZScore"], slope * df["NTL_ZScore"] + intercept, color="#ff0055", label=f"Linear Fit (r={r:.3f})")
plt.title("VIIRS NTL Radiance Anomalies vs GDP Growth (CZ)")
plt.xlabel("Nocturnal Radiance Anomaly (Z-Score)")
plt.ylabel("YoY GDP Growth (%)")
plt.grid(True, linestyle="--", alpha=0.5)
plt.legend()
plt.show()
```
