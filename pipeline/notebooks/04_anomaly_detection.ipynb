### Load ML Models

```python
import numpy as np
from pipeline.src.models import AnomalyDetector
print("Anomaly detection models successfully imported.")
```

### Run Unsupervised Isolation Forest on Features

```python
# Create synthetic pixel profiles
# D = [NDVI, NDWI, BSI]
np.random.seed(42)
normal_pixels = np.random.normal(loc=[0.6, -0.2, 0.2], scale=0.05, size=(90, 3))
anomaly_pixels = np.random.normal(loc=[-0.1, 0.1, 0.7], scale=0.05, size=(10, 3))
pixel_features = np.vstack([normal_pixels, anomaly_pixels])

detector = AnomalyDetector(contamination=0.10)
predictions, scores = detector.fit_predict_spatial(pixel_features)

print(f"Isolated {np.sum(predictions == -1)} anomalous pixels out of {len(predictions)} total pixels.")
```
